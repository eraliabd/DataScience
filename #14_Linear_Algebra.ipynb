```python
# Linear Algebra
from numpy.linalg import inv
```

```python
a = np.array([[1., 2.], [3., 4.]])
```

```python
a_inv = inv(a)
a_inv
```

### Agar a matritsa ob'ekti bo'lsa, qaytariladigan qiymat ham matritsadir:

```python
ainv = inv(np.matrix(a))
ainv
```
