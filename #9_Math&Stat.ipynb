### `sum` metodi

```python
arr.sum() # arr ning barcha elementlarni yig'indisini hisoblab beradi
```

```python
arr.sum(axis=0) # ustun yig'indi
```

### `mean` metodi

```python
# sodda ma'lumot
data = np.arange(5) # 0 ~ 5 gacha bo'lgan elementlarda tashkil topgan massiv
data
```

```python
mean = np.sum(data)/len(data)
mean
```

```python
# numpy mean
np.mean(data)
```

```python
data.mean()
```
