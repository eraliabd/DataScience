# **Ma'lumotlar tahlili. Pandas kutubxonasi**

## Foydali funksiyalar

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Yil': [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
    'Aholi soni': [33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 30],
    'Temp': [1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
    'Zichlik': [79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]
}

df = pd.DataFrame(data)
df
```

```python
df.index = [21, 20, 19, 18, 17, 16, 15, 10]
```

```python
df
```

`reindex`

```python
ind = [21, 20, 19, 18, 17, 16, 15, 10]
ind.sort()
ind
```

```python
df.reindex(ind)  # yangi dataframe yaratdi
```

```python
df
```

```python
df = df.reindex(ind)
df
```

```python
df.reindex([10, 15, 20])
```

```python
df.reindex(index=range(10, 22))
```

```python
df.reindex(index=range(10, 22), method='nearest')  # nearest metodi - bo'sh joylarni yaqin ma'lumot bilan to'ldiradi
```

```python
df.reindex(index=range(10, 22), method='ffill')  # forward fill - boshidagi qiymatni bo'sh joylarga yozib(nusxalab) ketadi
```

```python
df
```

```python
df2 = df.reindex(columns=['Aholi soni', 'Temp', 'Zichlik'])  # Yil ustunini tashlab yubordik
df2
```

```python
df2.head() # boshlang'ich 5 ta rowni qaytaradi
```

```python
df2.sample() # tasodifiy 1 ta qatorni qaytaradi, sample(3) - tasodidiy 3 ta
```

```python
df2.sample(3)  # tasodidiy 3 ta
```

```python
df.loc[[10, 15, 18, 20], ['Aholi soni', 'Temp', 'Zichlik']]  # shu tartibda ham ajratib olish mumkin
```

### Ma'lumotlarni tashlab yuborish
`drop` metodi

```python
df2
```

```python
df2.drop(10)  # 10-indexdagi qiymatni tashlab yuborish
```

```python
df
```

```python
df.drop([15, 20])  # bir nechta qatorlarni tashlab yuborish
```

```python
df.drop('Yil', axis=1)  # axis=1 - ustun, axis=0 - bu default ya'ni row(qator bilan ishlashni bildiradi)
```

```python
df_new = df.drop('Yil', axis=1)
df_new
```

```python
df.drop('Yil', axis=1, inplace=True)  # inplace=True - joyida o'zgartir degani ya'ni df=df.drop('Yil', axis=1) deb qayta o'zlashtirish shart emas.
# axis=1 ni o'rniga axis='columns' dan foydalansa ham bo'ladi
```

```python
df.drop([10, 15, 20], inplace=True)
```
