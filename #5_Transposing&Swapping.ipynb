### `reshape` metodi

```python
A = np.arange(6).reshape(3, 2)  # A massiv : Elementlari 0~6 gacha , shape (3, 2)
A
```

### `swaping axes`

```python
A.swapaxes(1, 0) # A massivning Transpose ko'rinishi beradi
```

### `Tranpose`ning ishlatilinishiga bir misol $A A^T$

```python
A
```

```python
np.dot(A, A.T) # AA^T hosil bo'lgan massiv
```
