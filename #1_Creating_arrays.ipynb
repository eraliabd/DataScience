**Ma'lumotlar tahlili. (NumPy kutubxonasi)**

**NumPy kutubxonasini chaqirib olish.**

```python
import numpy as np
```

**N-o'lchovli massiv(array)larga ishlov berish**



NumPyda massiv yaratishning eng oson yo'llaridan biri bu `array` funksiyasi yordamida amalga oshiriladi. Va ushbu funksiya har qanday turdagi ketma-ket ma'lumotlarni qabul qilib, uni yangi NumPy arrayiga o'girib beradi.

```python
data1 = [2, 5.6, 7, 9]
arr1 = np.array(data1)
```

```python
arr1
```

```python
data2 = (4, 5.6, 7, 9)
arr2 = np.array(data2)
```

```python
arr2
```

# ***Massiv o'lchami***

```python
arr1.ndim # array1 ning o'lchami
```

```python
arr2.ndim # array2 ning o'lchami
```

# **Ikki va undan katta o'lchamli massivlar**

```python
data3 = [[1,2,3,4], [5,6,7,8]] # list ichidagi list (nested list)
arr3 = np.array(data3) # array3
```

```python
data3
```

```python
arr3
```

```python
arr3.ndim # o'lchami
```

# **`shape`, `size` metodlari**



`shape` metodi massivning qator va ustunlar sonini ko'rsatadi, `size` esa massivlardagi elementlar sonini namoyon etadi.

```python
arr3.shape # qator va ustunlar sonini ko'rsatadi (qator, ustun)
```

```python
arr3.size # array3 ning elementlar soni
```

# N-o'lchamli massivlarni yaratishning boshqa usullari

`zeros` va `ones` funksiyalari yordamida massivlar yaratish

```python
arr4 = np.zeros(shape=(2, 4)) # barcha elementlari 0 ga teng bo'lgan (2, 4) o'lchamli massiv
arr4
```

```python
arr5 = np.ones(shape=(2, 4)) # barcha elementlari 0 ga teng bo'lgan (2, 4) o'lchamli massiv
arr5
```

`arange` funksiyasi yordamida yaratish

```python
arr6 = np.arange(4, 20, 2) # 4 dan 20 gacha bo'lgan qiymatlarni 2 qadam bilan massivga joylab beradi.
arr6
```

```python
arr7 = np.arange(20) # 0 dan 20 gacha bo'lgan qiymatlarni 1 qadam bilan massivga joylab beradi.
arr7
```
